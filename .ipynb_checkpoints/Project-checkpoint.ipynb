# 날짜형 정제

```python
def add_zero(date):
    year, month, day = date[:-1].split('.')
    if len(month) == 1:
        month = '0'+month
    if len(day) == 1:
        day = '0'+day
    return '.'.join([year, month, day])

data['date'] = pd.to_datetime(data['date'].apply(add_zero), format='%Y.%m.%d')
# for i in range(1,6):
#     globals()['rank'+str(i)]['date'] = globals()['rank'+str(i)]['date'].apply(add_zero)
#     globals()['rank'+str(i)]['date'] = pd.to_datetime(globals()['rank'+str(i)]['date'], format='%Y.%m.%d')
```
